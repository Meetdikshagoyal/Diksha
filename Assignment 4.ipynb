```python
def calculate_channeltype_distribution(df_raw):
    """
    Function to calculate the distribution of channel types from the top 1000 records.
    
    Parameters:
    data (DataFrame): DataFrame containing the YouTube channel data.
    
    Returns:
    dict: Dictionary containing the distribution of channel types.
    """
    top_1000_data = df_raw.head(1000)
    channeltype_distribution = top_1000_data['channeltype'].value_counts().to_dict()
    return channeltype_distribution
```

```python
# Call the function to calculate channel type distribution
distribution = calculate_channeltype_distribution(df_raw)
print(distribution)
```
